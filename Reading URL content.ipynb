```python
import requests

URL = 'https://github.com/scikit-image/skimage-tutorials'

#retrieving the data from the specific resource by using 'request.get'
response_object = requests.get(URL)
#getting the HTML text of the URL content
print(response_object.text)

```

```python
#status_code returns a number that indicates the status 
#(200 is OK, 404 is Not Found, those two are two most common status code )
print(response_object.status_code)
```

```python
#viewing the response headers
print(response_object.headers)
```

```python
#showing type of content of the header
print(response_object.headers['Content-Type'])
```
